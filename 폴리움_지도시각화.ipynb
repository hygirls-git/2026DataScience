```python
import folium
univ_map = folium.Map(
    location=[37.573050, 126.979189], tiles ='OpenStreetMap' , zoom_start=11 )

univ_map
```

```python
import csv
f=open('univ.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)
for row in data:
    # print(row[14], row[15])

    folium.CircleMarker(
        location=[row[14], row[15]],
        color='#FF0000',
        fill_color='#0000FF',
        fill_opacity=0.9,
        weight=1.5,
        radius=5
    ).add_to(univ_map)

univ_map
```
